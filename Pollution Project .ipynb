```python
import cv2
import numpy as np
import os
```

```python


x1, y1 = -1, -1
drawing = False
image = np.full((500, 500, 3), (235, 206, 135), dtype=np.uint8)
c = 0


def mouse_tracking(event, x, y, flags, param):
    global x1, y1, drawing, image
    if event == cv2.EVENT_LBUTTONDOWN: 
        drawing = True
        x1, y1 = x, y
    elif event == cv2.EVENT_MOUSEMOVE and drawing: 
        temp_image = image.copy() 
        cv2.rectangle(temp_image, (x1, y1), (x, y), (255, 255, 255), -1)
        cv2.imshow("pollution", temp_image)
    elif event == cv2.EVENT_LBUTTONUP:
        drawing = False
        cv2.rectangle(image, (x1, y1), (x, y), (255, 255, 255), -1)
    elif event == cv2.EVENT_RBUTTONDOWN: 
        cv2.circle(image, (x, y), 20, (42, 42, 139), -1) 
    elif event == cv2.EVENT_MBUTTONDOWN:  
        cv2.circle(image, (x, y), 15, (0, 0, 0), -1) 
    elif event == cv2.EVENT_LBUTTONDBLCLK: 
        cv2.circle(image, (x, y), 15, (0, 255, 0), -1)  
    elif event == cv2.EVENT_MOUSEWHEEL:  
        cv2.circle(image, (x, y), 9, (245, 245, 245), -1)  


cv2.namedWindow("pollution")
cv2.setMouseCallback("pollution", mouse_tracking)

while True:
   
    cv2.imshow("pollution", image)

    
    if cv2.waitKey(1) & 255 == ord("s"):
        cv2.imwrite(r"C:\Users\ADMIN\Downloads\space1\{}.jpg".format(c), image)
        print(f"Frame {c} has been captured")
        c += 1 

    
    if cv2.waitKey(1) & 255 == ord("r"):
        image = np.full((500, 500, 3), (235, 206, 135), dtype=np.uint8) 

    
    if cv2.waitKey(1) & 255 == ord("q"):
        break


cv2.destroyAllWindows()
```

```python
from PIL import Image

folder_path = r"C:\Users\ADMIN\Downloads\space1"
file_list = os.listdir(folder_path)
sorted_files = sorted(file_list, key=lambda x: int(x.split('.')[0]))
first_frame = True
frame_width, frame_height = 700, 500

frames = []
frame_count = 0

for file_name in sorted_files:
    file_path = os.path.join(folder_path, file_name)
    frame_count += 1

    img = cv2.imread(file_path)
    
    if img is not None:
        img = cv2.resize(img, (frame_width, frame_height))

        if first_frame:
            cv2.putText(img, "The Impact of Factory Pollution on Trees", (20, 40), 
                        cv2.FONT_HERSHEY_COMPLEX_SMALL, 0.9, (0, 100, 0), 1, cv2.LINE_AA)
            first_frame = False  

        if frame_count == 20:
            cv2.putText(img, "Pollution...!", (200, 300), 
                        cv2.FONT_HERSHEY_COMPLEX_SMALL, 2, (0, 0, 255), 1, cv2.LINE_AA)
        
        frame_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        frames.append(Image.fromarray(frame_rgb))
    else:
        print(f"Failed to load image: {file_name}")

description_page = np.zeros((frame_height, frame_width, 3), dtype=np.uint8) + 255
cv2.putText(description_page, "Project Title: The Impact of Factory Pollution on Trees", (50, 50), 
            cv2.FONT_HERSHEY_COMPLEX_SMALL, 1, (0, 0, 255), 1, cv2.LINE_AA)
cv2.putText(description_page, "Team Members:", (50, 100), 
            cv2.FONT_HERSHEY_COMPLEX_SMALL, 1, (0, 0, 0), 1, cv2.LINE_AA)
cv2.putText(description_page, "1. Sahitya.S", (50, 140), 
            cv2.FONT_HERSHEY_COMPLEX_SMALL, 1, (0, 0, 0), 1, cv2.LINE_AA)
cv2.putText(description_page, "2. Harshitha.M", (50, 180), 
            cv2.FONT_HERSHEY_COMPLEX_SMALL, 1, (0, 0, 0), 1, cv2.LINE_AA)
cv2.putText(description_page, "3. Jayasree.M", (50, 220), 
            cv2.FONT_HERSHEY_COMPLEX_SMALL, 1, (0, 0, 0), 1, cv2.LINE_AA)
cv2.putText(description_page, "4. Roopa.Ch", (50, 260), 
            cv2.FONT_HERSHEY_COMPLEX_SMALL, 1, (0, 0, 0), 1, cv2.LINE_AA)
cv2.putText(description_page, "Thank You!", (50, 320), 
            cv2.FONT_HERSHEY_COMPLEX_SMALL, 1, (0, 100, 0), 1, cv2.LINE_AA)

description_rgb = cv2.cvtColor(description_page, cv2.COLOR_BGR2RGB)
frames.append(Image.fromarray(description_rgb))

output_gif_path = r"C:\Users\ADMIN\Downloads\Pollution.gif"
frames[0].save(output_gif_path, save_all=True, append_images=frames[1:], duration=1000, loop=0)

print(f"GIF saved at {output_gif_path}")
cv2.destroyAllWindows()
```
